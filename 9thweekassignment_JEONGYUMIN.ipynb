lab 8-1: 오늘의 날짜와 현재 시간 출력하기

```python
from datetime import datetime
now=datetime.now()
year=now.year
month=now.month
day=now.day
hour=now.hour
minute=now.minute
second=now.second

if hour<12:
    am_pm='오전'
    display_hour=hour
else:
    am_pm='오후'
    display_hour=hour-12

if display_hour==0:
    display_hour=12

print(f'오늘의 날짜 :{year}년 {month}월 {day}일')
print(f'현재시간:{am_pm} {display_hour}시 {minute}분 {second}초')
```

Lab 8-2: 남은 날짜 계산하기

```python
#1)
from datetime import datetime
today=datetime.now()
print(f'오늘은{today.year}년 {today.month}월 {today.day}일입니다.')

christmas=datetime(2027,12,25,0,0,0)
diff=christmas-today
days=diff.days
hours=diff.seconds//3600
print(f'2027년 크리스마스 까지는 {days}일 {hours}시간 남았습니다.')
```

연습문제 8.1
datetime 모듈을 사용하여 연애 시작일로부터 오늘까지 며칠이 지났는지 계산하고 100일/200일/500일/1000일 기념일을 출력한다

```python
from datetime import date, timedelta

start_day = date(2019,2,24)

today=date.today()
passed_days=today-start_day

print(f'춘향이와 몽룡이의 연애 시작일:{start_day.year}년 {start_day.month}월 {start_day}일')
print(f'연애 시작일로부터 경과한 날짜:{passed_days.days}일')
```

```python
from datetime import date, timedelta

start_day = date(2019,2,24)
print(f'춘향이와 몽룡이의 연애 시작일:{start_day.year}년 {start_day.month}월 {start_day.day}일')

for day in [100,200,500,1000]:
    anniversary = start_day + timedelta(days=day-1)
    print(f'{day}일 기념일:{anniversary.year}년 {anniversary.month}월 {anniversary.day}일')
```

연습문제 8.2
1부터 1,000,000 까지 정수의 합을 구하는 함수를 작성하고 time모듈을 사용하여 실행 시간을 측정한다.

```python
import time

def sum1to1000000():
    total=0
    for i in range(1,1000001):
        total+=i
    return total
start=time.time()
sum1to1000000()
end = time.time()
print(f'1에서 1,000,000 까지의 합을 구하는 시간:{end-start:.5f}초')
```

```python
import time
def sum1to1000000():
    total=0
    for i in range(1,1000001):
        total+=i
    return total
start=time.time()

for i in range(100):
    sum1to1000000()
end =time.time()

print(f'1,000,000까지의 합을 100번 반복해서 구하는 시간:{end-start:.4f}초')
```

연습문제 8.3

여러 함수를 각각 100번 실행하고 time모듈을 사용하여 실행시간을 측정한다.

```python
import time
def hello():
    print('Hello Python!')
start=time.time()
for i in range(100):
    hello()

end=time.time()
print(f'Hello Python! 출력을 100번 반복하는 시간: {end-start:.4f}초')
```

```python
import time
import math

def factorial_1000():
    return math.factorial(1000)
start=time.time()
for i in range(100):
    factorial_1000()
end=time.time()
print(f'1000!을 100번 반복해서 구하는 시간:{end-start:.4f}초')
```

```python
import time
def odd_cube_sum():
    total=0
    for i in range(1,1001,2):
        total+=1**3
    return total
start=time.time()

for i in range(100):
    odd_cube_sum()
end=time.time()
print(f'1에서 1000까지 홀수의 세제곱 더하기를 100번 반복하는 시간 " {end-start:.4f}초')
```

```python
import time
import math
def sin_sum():
    total=0
    for degree in range(1,361):
        radian=math.radians(degree)
        total+=math.sin(radian)
    return total
start=time.time()
for i in range(100):
    sin_sum()
end=time.time()
print(f'1에서 360도까지 sin값의 합을 100번 반복하는 시간: {end-start:.4f}초')
```
